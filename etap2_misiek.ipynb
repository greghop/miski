```python
# zad 7
napis = "codeee<<fon<ria"

tablica_znakow = []
for litera in napis:   
    if litera != "<":
        tablica_znakow.append(litera)
    else:
        if len(tablica_znakow) > 0:
            nowa_talica = []          
            for index_tablicy in range(len(tablica_znakow)-1):
                nowa_talica.append(tablica_znakow[index_tablicy])

            tablica_znakow = nowa_talica

koncowy_napis = ""
for znaki in tablica_znakow:
    koncowy_napis += znaki

print(koncowy_napis)
```

```python
# zad  8
napis = "A B C D E F G H H H"
tablica_napis=napis.split()
dlugosc_tablicy=len(tablica_napis)

tablica_kontrolna = []
tablica_licznikow = []
for i in range(dlugosc_tablicy):
    licznik = 0

    for j in range(dlugosc_tablicy):
        if tablica_napis[i] == tablica_napis[j]:
            licznik += 1

    if tablica_napis[i] not in tablica_kontrolna:
        tablica_kontrolna.append(tablica_napis[i])
        tablica_licznikow.append(licznik)

print(tablica_licznikow)

tablica_sum = []
elementow = len(tablica_licznikow)
for i in range(1, elementow + 1):
    licznik = 0  
    for j in range(elementow):        
        if tablica_licznikow[j] >= i:
            licznik+=i

    tablica_sum.append(licznik)

max_liczba = 0
for liczba in tablica_sum:
    if liczba>max_liczba:
        max_liczba = liczba

print(max_liczba)

# 2 2 1 1 1 - 5
# 3 2 2 1 1 - 6
```

```python
# zad 9
napis = "--<>--"

len_napis = len(napis)

tablica_open_and_close = [0]
for i in range(len_napis):
    if napis[i] == "<" or napis[i] == ">":
        tablica_open_and_close.append(i)

if tablica_open_and_close[len(tablica_open_and_close)-1] != len_napis - 1:
    tablica_open_and_close.append(len_napis-1)


licznik = 0
for i in range(len(tablica_open_and_close)-1):
    
    start = napis[tablica_open_and_close[i]]
    stop = napis[tablica_open_and_close[i+1]]
    
    if start == "-" and stop == ">":
        licznik = licznik + tablica_open_and_close[i+1] - tablica_open_and_close[i]
    elif start == "<" and stop == ">":
        licznik = licznik + 2*(tablica_open_and_close[i+1] - tablica_open_and_close[i] - 1)
    elif start == ">" and stop == ">":
        licznik = licznik + 1*(tablica_open_and_close[i+1] - tablica_open_and_close[i] - 1)
    elif start == "<" and stop == "<":
        licznik = licznik + 1*(tablica_open_and_close[i+1] - tablica_open_and_close[i] - 1)
    elif start == "<" and stop == "-":
        licznik = licznik + 1*(tablica_open_and_close[i+1] - tablica_open_and_close[i] - 1)

print(licznik)
```
